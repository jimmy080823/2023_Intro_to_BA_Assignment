```python
def sigmoid(x):
    s = 1 / (1+np.exp(-x))
    d = 0.5*s
    f = d-0.25
    return f

x = np.linspace(-5, 5, 100)
ty = np.tanh(x)

plt.plot(x, ty, label='tanh')
plt.legend()
plt.plot(x, sigmoid(x), label='sigmoid')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 創建一個 x 範圍
x = np.linspace(-5, 5, 100)

# 計算tanh和sigmoid函數的值
tanh_y = np.tanh(x)
sigmoid_y = 1 / (1 + np.exp(-x))

# 調整sigmoid函數，使其經過(0, 0.5)和(-4, 0)
sigmoid_y = 0.5 * sigmoid_y
sigmoid_y = sigmoid_y - 0.25  # 平移函數以使其經過(-4, 0)

# 創建一個圖表
plt.figure(figsize=(10, 5))

# 繪製tanh函數
plt.plot(x, tanh_y, label='tanh', color='b')
plt.title('tanh 函數')
plt.grid(True)
plt.legend()

# 繪製sigmoid函數
plt.plot(x, sigmoid_y, label='sigmoid', color='r')
plt.title('Sigmoid 函數')
plt.grid(True)
plt.legend()

# 顯示圖表
plt.tight_layout()
plt.show()
```
